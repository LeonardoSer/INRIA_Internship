```python
YEARS = get_int_years_event()
COLLABORATION_EVENTS = get_collabs_event()
AUTHOR_EVENTS = get_auths_event()
```

```python
plt.figure(figsize=(15, 7), dpi=80)
plt.title("Distribution of new collaborations by year")
plt.xlabel("year")
plt.ylabel("# new_collaborations")
plt.xticks(rotation=90)

prev = 0
for y in YEARS:
    if(y!=1990):
        prev = COLLABORATION_EVENTS[YEARS.index(y-1)]
    num_new_collabs = COLLABORATION_EVENTS[YEARS.index(y)] - prev
    plt.scatter(y,  num_new_collabs, label=str(y) + " -> " + str(num_new_collabs))
    
plt.legend(ncol=2)
plt.xticks(YEARS)
plt.show()
```

```python
plt.figure(figsize=(15, 7), dpi=80)
plt.title("Distribution of new authors by year")
plt.xlabel("year")
plt.ylabel("# new_authors")
plt.xticks(rotation=90)

prev = 0
for y in YEARS:
    if(y!=1990):
        prev = AUTHOR_EVENTS[YEARS.index(y-1)]
    num_new_auths = AUTHOR_EVENTS[YEARS.index(y)] - prev
    plt.scatter(y,  num_new_auths, label=str(y) + " -> " + str(num_new_auths))
    
plt.legend(ncol=2)
plt.xticks(YEARS)
plt.show()
```

# 2.2.6

```python
plt.figure(figsize=(15, 5), dpi=80)
plt.title("Ratio between the number of new collaborations and new authors")
plt.xlabel("years")
plt.ylabel("#_new_collaborations / #_new_authors")
plt.xticks(rotation=90)

prev_auth = 0
prev_coll = 0
ratios = []
labels = []
for y in YEARS:
    if(y!=1990):
        prev_auth = AUTHOR_EVENTS[YEARS.index(y-1)]      
        prev_coll = COLLABORATION_EVENTS[YEARS.index(y-1)]
    num_new_collabs = COLLABORATION_EVENTS[YEARS.index(y)] - prev_coll
    num_new_auths = AUTHOR_EVENTS[YEARS.index(y)] - prev_auth
    ratio = num_new_collabs/num_new_auths
    labels.append(str(y) + " -> " + str('{:.2f}'.format(ratio)))
    ratios.append(ratio)

plt.plot(YEARS, ratios, color="black")
    
plt.xticks(YEARS)
plt.show()

plt.figure(figsize=(15, 5), dpi=80)
plt.title("Ratio between the number of new collaborations and new authors stretched by new collaboration event")
plt.xlabel("years stretched by collaborations events")
plt.ylabel("#_new_collaborations / #_new_authors")
plt.xticks(rotation=90)

prev_auth = 0
prev_coll = 0
ratios = []
labels = []
for y in YEARS:
    if(y!=1990):
        prev_auth = AUTHOR_EVENTS[YEARS.index(y-1)]      
        prev_coll = COLLABORATION_EVENTS[YEARS.index(y-1)]
    num_new_collabs = COLLABORATION_EVENTS[YEARS.index(y)] - prev_coll
    num_new_auths = AUTHOR_EVENTS[YEARS.index(y)] - prev_auth
    ratio = num_new_collabs/num_new_auths
    
    labels.append(str(y) + " -> " + str('{:.2f}'.format(ratio)))
    ratios.append(ratio)

plt.plot(COLLABORATION_EVENTS[1:], ratios, color="orange")
    
plt.xticks(COLLABORATION_EVENTS[1:], YEARS)
plt.show()


plt.figure(figsize=(15, 5), dpi=80)
plt.title("Ratio between the number of new collaborations and new authors stretched by new author event")
plt.xlabel("years stretched by author events")
plt.ylabel("#_new_collaborations / #_new_authors")
plt.xticks(rotation=90)

prev_auth = 0
prev_coll = 0
ratios = []
labels = []
for y in YEARS:
    if(y!=1990):
        prev_auth = AUTHOR_EVENTS[YEARS.index(y-1)]      
        prev_coll = COLLABORATION_EVENTS[YEARS.index(y-1)]
    num_new_collabs = COLLABORATION_EVENTS[YEARS.index(y)] - prev_coll
    num_new_auths = AUTHOR_EVENTS[YEARS.index(y)] - prev_auth
    ratio = num_new_collabs/num_new_auths
    
    labels.append(str(y) + " -> " + str('{:.2f}'.format(ratio)))
    ratios.append(ratio)

plt.plot(AUTHOR_EVENTS[1:], ratios, color="blue")
    
plt.xticks(AUTHOR_EVENTS[1:], YEARS)
plt.show()
```

```python
plt.figure(figsize=(15, 5), dpi=80)
plt.title("Ratio between the total number of collaborations and authors")
plt.xlabel("years")
plt.ylabel("#_collaborations / #_authors")
plt.xticks(rotation=90)


ratios = []
labels = []
for y in YEARS:
    num_new_collabs = COLLABORATION_EVENTS[YEARS.index(y)] 
    num_new_auths = AUTHOR_EVENTS[YEARS.index(y)] 
    ratio = num_new_collabs/num_new_auths
    labels.append(str(y) + " -> " + str('{:.2f}'.format(ratio)))
    ratios.append(ratio)

plt.plot(YEARS, ratios, color="black")
    
plt.xticks(YEARS)
plt.show()

plt.figure(figsize=(15, 5), dpi=80)
plt.title("Ratio between the total number of collaborations and authors stretched by new collaboration event")
plt.xlabel("years stretched by collaborations events")
plt.ylabel("#_collaborations / #_authors")
plt.xticks(rotation=90)

ratios = []
labels = []
for y in YEARS:
    num_new_collabs = COLLABORATION_EVENTS[YEARS.index(y)] 
    num_new_auths = AUTHOR_EVENTS[YEARS.index(y)] 
    ratio = num_new_collabs/num_new_auths
    
    labels.append(str(y) + " -> " + str('{:.2f}'.format(ratio)))
    ratios.append(ratio)

plt.plot(COLLABORATION_EVENTS[1:], ratios, color="orange")
    
plt.xticks(COLLABORATION_EVENTS[1:], YEARS)
plt.show()


plt.figure(figsize=(15, 5), dpi=80)
plt.title("Ratio between the total number of collaborations and authors stretched by new author event")
plt.xlabel("years stretched by author events")
plt.ylabel("#_collaborations / #_authors")
plt.xticks(rotation=90)

ratios = []
labels = []
for y in YEARS:
    num_new_collabs = COLLABORATION_EVENTS[YEARS.index(y)] 
    num_new_auths = AUTHOR_EVENTS[YEARS.index(y)] 
    ratio = num_new_collabs/num_new_auths
    
    labels.append(str(y) + " -> " + str('{:.2f}'.format(ratio)))
    ratios.append(ratio)

plt.plot(AUTHOR_EVENTS[1:], ratios, color="blue")
    
plt.xticks(AUTHOR_EVENTS[1:], YEARS)
plt.show()
```

```python
import numpy as np

colls = get_all_collabs()["2018"].values

unique, counts = np.unique(colls, return_counts=True)
    
plt.figure(figsize=(15, 5), dpi=80)
plt.title("Degree distribution (Distribution on the number of collaborations per author)")
plt.xlabel("#_collaborations")
plt.ylabel("#_authors")
plt.xticks(rotation=90)

plt.plot(unique, counts, color="blue")
    
plt.show()

plt.figure(figsize=(15, 5), dpi=80)
plt.title("log-log form of the degree distribution (Distribution on the number of collaborations per author)")
plt.xlabel("log(#_collaborations)")
plt.ylabel("log(#_authors)")
plt.xticks(rotation=90)

plt.plot( np.log(unique), np.log(counts), color="orange")
    
plt.show()
```
