```python
from myLibraries.queries import * 
from myLibraries.events import *
import matplotlib.pyplot as plt

# function for granting data (focal_controls)
focal_control = get_all_granting_data() # get all data about focal and controls with the respective collaboration data
granted = get_granted() # get just the granted data
not_granted = get_not_granted() #get just the not granted data

focal_control
```

```python
# given the group ID return the associated focal and control IDs, trajectories and granting YEAR
group_id = 6
focal_ID, control_ID, y_focal, y_control, grant_year = get_focal_control_traj_byGroup(group_id)
```

```python
## Plot group trajectories
plt.figure(figsize=(10, 6), dpi=80)
plt.title("focal-control trajectories for group " + str(group_id) + " with granting year " + grant_year +"\n", fontdict = {'fontsize' : 12})
plt.xlabel("#new_collabs as event")
plt.ylabel("#num_Collaborations")
plt.xticks(rotation='vertical')

plt.plot(EVENTS, y_focal, label="focal_ID -> " + focal_ID)
plt.plot(EVENTS, y_control, label="control_ID -> " + control_ID)

plt.legend(loc='upper left')
plt.xticks(EVENTS, YEARS)
plt.show()
```

```python
# Average granted/not_granted trajectories by start year
EVENTS = get_collabs_event()
start_y = "2000"
focal_x, focal_avg = get_focals_avg_trajectories(EVENTS, start_y) 
control_x, control_avg = get_controls_avg_trajectories(EVENTS, start_y)
```

```python
## plot avg trajectories by starting year
plt.figure(figsize=(25, 8), dpi=80)
plt.title("average trajectories for starting year"+ start_y+" \n", fontdict = {'fontsize' : 12})
plt.xlabel("#new_collabs as event")
plt.ylabel("#num_Collaborations")
plt.xticks(rotation='vertical')

plt.plot(focal_x, focal_avg,label="focal")
plt.plot(control_x, control_avg, label="control")

plt.legend(loc='upper left')
plt.xticks(EVENTS, YEARS)
plt.show()
```
